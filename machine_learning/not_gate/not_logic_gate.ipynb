```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0],[1]])
y = np.array([[1],[0]])
print(x)
print(y)
```

```python
def ReLU(x):
    return np.maximum(0, x)

# Derivada de la funciÃ³n ReLU para arrays de numpy
def deriv_ReLU(x):
    return np.where(x >= 0, 1, 0)

def MSE(y, a_2):
    return 0.5 * np.sum((a_2 - y)**2)
```

```python
# initializa params randomly
w_1 = np.random.rand()
w_2 = np.random.rand()
b_1 = np.random.rand()
b_2 = np.random.rand()

w_1
```

```python
Cs = []
iterations = [i for i in range(10000)]
learning_rate1 = 0.001
for i in iterations:
    # forward propagation
    z_1 = w_1 * x + b_1
    a_1 = ReLU(z_1)
    z_2 = w_2 * a_1 + b_2
    a_2 = ReLU(z_2)
    # gradient_descent
    delta_2 =(a_2 - y) * deriv_ReLU(z_2)
    delta_1 = w_2 * deriv_ReLU(z_1)
    partial_w2 = delta_2 * a_1
    partial_b2 = delta_2 * 1
    partial_w1 = delta_1 * delta_2 * x
    partial_b1 = delta_1 * delta_2 * 1
    C = MSE(y, a_2)
    Cs.append(C)
    w_2 -= partial_w2 * learning_rate1
    b_2 -= partial_b2 * learning_rate1
    w_1 -= partial_w1 * learning_rate1
    b_1 -= partial_b1 * learning_rate1
    
```

```python
# 
plt.figure(figsize=(8, 6))
plt.plot(iterations, Cs, label='Sine wave', color='blue')
plt.title('MSE error')
plt.xlabel('n iterations')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
```
