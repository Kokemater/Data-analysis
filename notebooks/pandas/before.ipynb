```python
import pandas as pd
import numpy as np

# Definir alturas y días del mes
alturas = ['Altura1', 'Altura2', 'Altura3']
dias_del_mes = [f'Día{i+1}' for i in range(30)]

# Generar datos aleatorios de temperaturas
np.random.seed(0)  # Para reproducibilidad
data = {
    altura: np.random.uniform(low=-10, high=35, size=len(dias_del_mes))
    for altura in alturas
}

# Crear el DataFrame
df = pd.DataFrame(data, index=dias_del_mes)

# Calcular la media de las temperaturas para cada altura
df.loc['Media'] = df.mean(axis=0)

# Mostrar el DataFrame
df
```
